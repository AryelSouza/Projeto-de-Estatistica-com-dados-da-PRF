Para usar apenas o mês precisamos tratar a coluna data_inversa. Por isso importei datetime,onde podemos usar ferramentas para arrumar os dados.após utilizar a função pd.to_datetime() a coluna data_inversa sera mudada para uma data reconhecida pelo script. após posso colocar a função dt.month que identifica o mês da data.

```python
#Ano de 2018
df18['mes'] = pd.to_datetime(df18['data_inversa']).dt.month
mes18 = df18.loc[:,['mes','pessoas','ilesos','feridos_graves']]
c = mes18.groupby(by='mes').sum()
pr_ilesos = (c['ilesos']/c['pessoas'])
pr_feridos = (c['feridos_graves']/c['pessoas'])
prop = pd.DataFrame(pr_ilesos).rename(columns={0:'proporção ilesos'}).round(2)
prop['proporção feridos graves'] = pd.DataFrame(pr_feridos).round(2)
prop
```

Aqui podemos usar o drop para remover o 'index' e usar o groupby para agrupar os dados de antes e durante a pandemia

```python
dfpand = pd.concat([df18,df19,df20,df21], keys = ['2018','2019','2020','2021'],names=['ano','index'])
dfpand = dfpand.reset_index().drop(['index'],axis=1)
dfpand = dfpand.groupby(['ano']).count().iloc[:,:1].rename(columns={'id':'n° acidentes'})
antes = pd.DataFrame(dfpand.iloc[:2,:].sum()).rename(index={'n° acidentes':'antes da pandemia'})
durante = pd.DataFrame(dfpand.iloc[2:,:].sum()).rename(index={'n° acidentes':'durante a pandemia'})
pd.concat([antes,durante]).rename(columns={0:'acidentes'})
```
